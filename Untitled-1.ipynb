- Sex: This is a binary categorical column, so I will use label encoding (1 for female and 0 for male).
- Embarked: This is a categorical column with 3 possible values, so one-hot encoding is a good option.

```python
# Label encoding for 'Sex' column (male -> 0, female -> 1)
data['Sex'] = data['Sex'].map({'male': 0, 'female': 1})

# One-hot encoding for 'Embarked' column (create binary columns for each category)
data = pd.get_dummies(data, columns=['Embarked'], drop_first=True)
```

### 4. Conclusion
- Questions to answer?
1. Did children have a higher chance of survival?  

- Findings

 - Children (aged under 12) had a higher chance of survival compared to adults.

 - This could be attributed to the prioritization of children in the lifeboat boarding process, as they were likely considered more vulnerable during the chaos of the sinking.

2. Were women and children prioritized in lifeboat boarding?  

- Findings

- Women had a higher chance of survival compared to men.

- Women in all classes, including third class, had significantly higher survival rates.

 - The 'Women and Children First' protocol may have influenced this pattern.
 
3. Did passengers from higher classes have a higher chance of survival?  

- Finding: 

- Passengers in higher classes (first and second class) had a much higher chance of survival compared to those in third class. 

- First-class passengers had the highest survival rates.

- The disparity in survival rates across classes likely reflects socio-economic factors, such as access to lifeboats, proximity to the upper decks, and faster response times during the evacuation

4. How did age, sex and Embarkation Point  influence survival chances? 

- Findings

- Age: Younger passengers had a higher survival rate. Children were prioritized, but adults, especially men, had lower survival rates.

- Sex: Females had a higher chance of survival than males, which supports the hypothesis that women and children were given priority during lifeboat boarding.

- Embarkation Point: Passengers who embarked at Southampton had a lower chance of survival compared to those embarking at Cherbourg or Queenstown.
- This might reflect different boarding procedures, social dynamics, or ship-specific evacuation strategies.

5. Safety Measures and Emergency Preparedness: Use the insights from the Titanic disaster to inform modern safety strategies and policies in transport or public services.

- The analysis of the Titanic disaster has provided valuable lessons on the importance of prioritizing vulnerable groups, ensuring equitable access to emergency resources, and addressing socio-economic inequalities in emergency situations. Modern safety strategies should be informed by these lessons to improve evacuation procedures in case of disasters in transport or public services.

#### 5. RECOMMENDATIONS

- Based on the findings, the following recommendations can be made to improve safety measures and emergency preparedness in modern transport and public services:

1. Prioritize Vulnerable Populations:

 - Emergency evacuation plans should prioritize vulnerable groups, including children, elderly passengers, and those with disabilities. 

 - This can be achieved by designating specific lifeboats or evacuation routes for these groups to ensure faster access to safety.

2. Gender-Neutral Evacuation Policies:

- Modern evacuation strategies should not prioritize one gender over another, as it could lead to inequities during a crisis.
Instead, a focus on vulnerability (e.g., children, the elderly, and sick individuals) should guide decisions about who gets access to

safety resources first.

3. Ensure Equal Access to Lifeboats and Safety Equipment:

- Socio-economic class should not influence access to emergency resources. Ensuring that all passengers have equal access to lifeboats, regardless of their class, will increase the fairness and effectiveness of emergency evacuation procedures.

4. Improve Emergency Training and Preparedness for All Passengers:

- Passenger education programs should emphasize emergency preparedness and evacuation protocols.

 - This ensures that all passengers, regardless of background, are well-prepared to respond calmly and effectively in case of an 
 
 emergency.
